# Initial Data Exploration and Cleaning

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
#load and initially visualize the data
original_data = pd.read_csv('data.csv')
original_data.head()
```

```python
#initial counts of nan values
print('# of NaN values:')
print(original_data.isnull().sum())
```

```python
#remove unnamed 32 column because it contains NaN values
original_data.drop(columns=['Unnamed: 32'], inplace=True)
```

```python
#redisplay data without unnamed:32 column
original_data.head()
```

```python
#remove id bc it can't be used for classification  
original_data.drop(columns=['id'], inplace=True)
```

```python
#convert malignant=1 and benign=0 
original_data['diagnosis'] = original_data['diagnosis'].map({'M':1,'B':0})
original_data.head()
```

```python
#understand the range of data in dataset
original_data.describe()
```

```python
#split dataframe based on diagnosis (malignant=1 and benign=0)
malignant_data=original_data[original_data['diagnosis'] ==1]
benign_data=original_data[original_data['diagnosis'] ==0]
```

```python
# create a features df without the diagnosis
features = original_data.columns[1:]

#loop through each feature and create a plot
for feature in features:
    plt.figure(figsize=(8, 6))
    sns.histplot(malignant_data[feature], color='red', label='Malignant', kde=True)
    sns.histplot(benign_data[feature], color='blue', label='Benign', kde=True)
    plt.title(f'Distribution of {feature} for Benign and Malignant Tumors')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```

```python
from sklearn.utils import resample
from sklearn.model_selection import train_test_split

# Split data with stratified sampling
X_train, X_val, y_train, y_val = train_test_split(original_data.drop(columns=['diagnosis']), original_data['diagnosis'], test_size=0.2, random_state=42, stratify=original_data['diagnosis'])

# Upsample the minority class in the training set
train_data = pd.concat([X_train, y_train], axis=1)
malignant_samples = train_data[train_data['diagnosis'] == 1]
benign_samples = train_data[train_data['diagnosis'] == 0]

# downsample malignant samples to address uneven sample count
benign_downsampled = resample(benign_samples, replace=True, n_samples=len(malignant_samples), random_state=42)

# Combine majority and upsampled minority class
train_balanced = pd.concat([benign_downsampled, malignant_samples])

# Separate features and target in the balanced dataset
X_train_balanced = train_balanced.drop(columns=['diagnosis'])
y_train_balanced = train_balanced['diagnosis']
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score

# Initialize models
logistic_regression = LogisticRegression()

# Train models
logistic_regression.fit(X_train_balanced, y_train_balanced)

# Predictions
y_pred_lr = logistic_regression.predict(X_val)

# Evaluate the model
print("Logistic Regression:")
print("Accuracy:", accuracy_score(y_val, y_pred_lr))
```

```python
from sklearn import tree

x, y = X_train_balanced, y_train_balanced
clf = tree.DecisionTreeClassifier(max_leaf_nodes = 6, criterion = 'entropy')
clf = clf.fit(x, y)

tree.plot_tree(clf)
plt.show()
```

```python
#evaluate tree model performance on validation set
from sklearn.metrics import classification_report, accuracy_score

# Predictions
y_pred_dt = clf.predict(X_val)

print("Accuracy:", accuracy_score(y_val, y_pred_dt))
```

```python
#another decision tree
plt.figure(figsize=(20,10))
tree.plot_tree(clf, feature_names=X_train_balanced.columns, class_names=['Benign', 'Malignant'], filled=True)
plt.show()
#by looking at this tree we can see that the perimeter is the most important feature in determining whether a tumor 
#is benign or malignant
```

```python
#bootstrap aggregating technique
from sklearn.tree import DecisionTreeClassifier
#define the number of trees to create
n_estimators = 10
#initialize
base_models = []

#loop through each base model
for _ in range(n_estimators):
    #randomly sample with replacement from the training data
    bootstrap_sample = resample(X_train, y_train, replace=True, random_state=42)
    X_bootstrap, y_bootstrap = bootstrap_sample
    
    #initialize a decision tree classifier as the base model
    base_model = DecisionTreeClassifier()
    
    #train the base model on the bootstrap sample
    base_model.fit(X_bootstrap, y_bootstrap)
    
    #add to the list
    base_models.append(base_model)

#predict validation set
predictions = np.array([model.predict(X_val) for model in base_models])
final_predictions = np.mean(predictions, axis=0)

#convert to 0 or 1
final_predictions = np.round(final_predictions).astype(int)

#calculate accuracy on the validation set
accuracy = accuracy_score(y_val, final_predictions)
print("Bagging Accuracy:", accuracy)
```

```python
#random forest model
from sklearn.ensemble import RandomForestClassifier

#initialize the classifier
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)

#train the Random Forest model on the balanced training data
random_forest.fit(X_train_balanced, y_train_balanced)

#predict
y_pred_rf = random_forest.predict(X_val)

print("Accuracy:", accuracy_score(y_val, y_pred_rf))

feature_importance = random_forest.feature_importances_

#sort feature importances in descending order
indices = np.argsort(feature_importance)[::-1]

#plot
plt.figure(figsize=(15, 10))
plt.title("Feature Importance")
plt.bar(range(X_train_balanced.shape[1]), feature_importance[indices], align="center")
plt.xticks(range(X_train_balanced.shape[1]), X_train_balanced.columns[indices], rotation=90)
plt.xlim([-1, X_train_balanced.shape[1]])
plt.show()
```
